```python
# notebooks/01_data_exploration.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('../data/raw/vehicles.csv')

print("Dataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing values:")
print(df.isnull().sum())
# Data exploration for Kenyan market conversion
import sys
sys.path.append('../src')
from data_preprocessing import DataPreprocessor
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize preprocessor
preprocessor = DataPreprocessor()

# Load and convert data
print("Loading and converting data to Kenyan market...")
df_kenya = preprocessor.load_and_clean_data('../data/raw/vehicles.csv')

print(f"Kenyan dataset shape: {df_kenya.shape}")
print(f"Price range in KES: {df_kenya['price_kes'].min():,.0f} - {df_kenya['price_kes'].max():,.0f}")

# Visualize price distribution in Kenyan market
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
df_kenya['price_kes'].hist(bins=50)
plt.title('Car Price Distribution in KES')
plt.xlabel('Price (KES)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
df_kenya['odometer_km'].hist(bins=50)
plt.title('Mileage Distribution in KM')
plt.xlabel('Odometer (KM)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Save exploration results
df_kenya.to_csv('../data/processed/kenyan_cars_exploration.csv', index=False)
```
